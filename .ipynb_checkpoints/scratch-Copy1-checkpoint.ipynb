```python
import numpy as np
from sklearn.svm import SVC
import os
from pathlib import Path
from matplotlib import pyplot as plt
from keras.preprocessing import image
```

# Data Preparation

```python
pd = Path('./Images/')
```

```python
print(pd,type(pd))
```

```python
dirs = pd.glob('*')
```

```python
classes_dict = {}
num_name = {}
image_data = []
image_labels = []
count = 0 

for folder_name in dirs:
    print(folder_name)
    name = str(folder_name).split('/')[-1]
    classes_dict[name] = count
    num_name[count] = name
    count+=1
    
    for img in folder_name.glob('*.jpg'):
        data = image.load_img(img,target_size=(50,50))
        img_array = image.img_to_array(data)
#         print(len(img_array))
        image_data.append(img_array)
        image_labels.append(classes_dict[name])
```

```python
print(len(image_labels),len(image_data))
image_data = np.array(image_data,dtype='float32')/255.0
image_labels = np.array(image_labels)
print(image_data.shape,image_labels.shape)
```

```python
print(classes_dict)
```

# shuffling data

```python
import random 

combined = list(zip(image_data,image_labels))
random.shuffle(combined)

image_data[:],image_labels[:] = zip(*combined)
print(image_data.shape,image_labels.shape)
```

```python
class classification:    
    def __init__(self,cl1,cl2):
        self.class1 = cl1
        self.class2 = cl2
        self.w = None
        self.b = None
        self.los = None
        
    def hypothesis(self,x,w,b,label):
        return label*(np.dot(w,x)+b)


    def loss(self,x,y,w,b,c):
        loss=0
        for ix in range(x.shape[0]):
            hp = self.hypothesis(x[ix],w,b,y[ix])
            loss+=max(1-hp,0)
        w2norm = np.dot(w,w)
        loss = (w2norm/2) + c*loss
        return loss
    
    def predict(self,x):
        temp = np.dot(self.w,x)+self.b
        if(temp>=0):
            return self.class1
        else:
            return self.class2

    def gradient(self,x,y,w,b,c):
        wgr = np.zeros(w.shape)
        bgr = 0
        for ix in range(x.shape[0]):
            ti = self.hypothesis(x[ix],w,b,y[ix])
            if(ti>=1):
                continue
            else:
                wgr += (-1)*y[ix]*x[ix]
                bgr += (-1)*y[ix]
        wgr = w + c*(wgr)
        bgr = c*bgr
        return wgr,bgr

    def gradientdescent(self,x,y,c=1,lr=0.00001):
        self.w = np.zeros((x.shape[1]))
        self.b = 0
        self.los = []
        
        for ix in range(500):
            wgr,bgr = self.gradient(x,y,self.w,self.b,c)
         
            self.w = self.w - lr*wgr
            self.b = self.b - lr*bgr
            self.los.append(self.loss(x,y,self.w,self.b,c))
    
#         print(np.unique(self.w))
#         print(self.b)
    
```

```python
llist = []
image_data = image_data.reshape((image_data.shape[0],-1))
print(image_data.shape)
n_classes = len(classes_dict)
for ix in range(n_classes-1):
    for mx in range(ix+1,n_classes):
        print(ix,mx)
        x_data = []
        y_data = []
        
        for tx in range(image_data.shape[0]):
            if image_labels[tx]==ix or image_labels[tx]==mx:
                x_data.append(image_data[ix])
                if(image_labels[tx]==ix):
                    y_data.append(+1)
                else:
                    y_data.append(-1)
        x_data = np.array(x_data)
        y_data = np.array(y_data)
#         print(x_data.shape,y_data.shape)
        pp = classification(ix,mx)
#         print(x_data.shape,y_data.shape)
        pp.gradientdescent(x_data[:,:],y_data[:])
        
        print(pp.los)
        plt.figure()
        plt.plot(pp.los)        
        plt.show()
        llist.append(pp)  
    
```

```python
def finalprediction(x,llist):
    count = []
#     print(x.shape)
    for ix in llist:
        temp = ix.predict(x)
        count.append(temp)
    
    count = np.array(count)
#     print(count)
    idx,val = np.unique(count,return_counts=True)
    max_idx = np.argmax(val)
    return idx[max_idx]
    
```

```python
print(num_name[finalprediction(image_data[1,:],llist)])
print(num_name[image_labels[1]])
```

```python
def accuracy(x,y,llist):
    acc = 0 
    for ix in range(x.shape[0]):
        pred = finalprediction(x[ix],llist)
        if(pred==y[ix]):
            acc+=1
        if(ix%50==0):
            print(acc)
    return acc/x.shape[0]
```

```python
print(accuracy(image_data,image_labels,llist))
```
